```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import patches
```

```python
data = pd.read_csv('frozen.csv')
```

```python
frozen = list(set(list(data['V1']) + list(data['V2'])))
len(frozen)
```

```python
# Create points for each world cup finalist
points = dict()
for i in range(len(frozen)):
    points[frozen[i]] = [i,0,1]
```

```python
points
```

```python
def calc_arc(x1, x2, y1, y2):
    mxmy = mx, my = [(x1 + x2) / 2, (y1 + y2) / 2]
    r = np.sqrt((x1 - mx)**2 + (y1 - my)**2)
    width = 2 * r
    height = 2 * r
    start_angle = np.arctan2(y1 - my, x1 - mx) * 180 / np.pi
    end_angle = np.arctan2(my - y2, mx - x2) * 180 / np.pi
    arc_params = [mxmy, width, height, start_angle, end_angle]
    return arc_params



# calculate the arc
arcs = []
dialogs = []
for i in range(len(data)):
    v1 = data['V1'][i]
    v2 = data['V2'][i]
    x1, y1 = points[v1][0], points[v1][1]
    x2, y2 = points[v2][0], points[v2][1]
    mxmy = mx, my = [(x1 + x2) / 2, (y1 + y2) / 2]
    r = np.sqrt((x1 - mx)**2 + (y1 - my)**2)
    width = 2 * r
    height = 2 * r
    start_angle = np.arctan2(y1 - my, x1 - mx) * 180 / np.pi
    end_angle = np.arctan2(my - y2, mx - x2) * 180 / np.pi
    dialog = min(v1,v2) + max(v1,v2)
    dialogs.append(dialog)
    linewidth = dialogs.count(dialog) ** 2
    arc = patches.Arc(mxmy, width, height, start_angle, end_angle, linewidth=linewidth)
    arcs.append(arc)
    
fig, ax = plt.subplots(1,1, figsize=(10, 10))
for arc in arcs:
    ax.add_patch(arc)


ax.set_xticks([x for x in range(len(frozen))])
ax.set_xticklabels(frozen,rotation=90)
ax.set_yticks([x for x in range(len(frozen))])
ax.set_yticklabels([])
ax.set_xlim(-0.1, 12.1)
ax.set_ylim(-0.1, 12.1)
plt.show()
```

```python
characters = list(set(list(data['V1']) + list(data['V2'])))
characters
```

```python
arcs = []

for index, row in data.iterrows():
    v1, v2 = row
    x1 = characters.index(v1)
    x2 = characters.index(v2)

    # calculate the arc
    mxmy = mx, my = [(x1 + x2) / 2, 0]
    r = np.sqrt((x1 - mx)**2 + (0 - my)**2)
    width = 2 * r
    height = 2 * r
    start_angle = np.arctan2(0 - my, x1 - mx) * 180 / np.pi
    end_angle = np.arctan2(my - 0, mx - x2) * 180 / np.pi

    arc = patches.Arc(mxmy, width, height, start_angle, end_angle)
    arcs.append(arc)


fig, ax = plt.subplots(1,1)
for arc in arcs:
    ax.add_patch(arc)

ax.set_xticks(list(range(len(characters))))
ax.set_xticklabels(characters,rotation=45)
ax.set_xlim(-0.1, 13.1)
ax.set_ylim(-0.1, 6)
plt.show()
```
